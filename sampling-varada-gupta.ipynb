```python
class0, class1 = df['Class'].value_counts()


class_0 = df[df['Class'] == 0]
class_1 = df[df['Class'] == 1]
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
underclass0 = class_0.sample(class1)

test_under = pd.concat([underclass0, class_1], axis=0)

print("total class of 1 and 0:",test_under['Class'].value_counts())# plot the count after under-sampeling
test_under['Class'].value_counts().plot(kind='bar', title='Target Count Values')
```

```python
overclass1 = class_1.sample(class0, replace=True)

test_over = pd.concat([overclass1, class_0], axis=0)

print("total class of 1 and 0:",test_under['Class'].value_counts())# plot the count after under-sampeling
test_over['Class'].value_counts().plot(kind='bar', title='Target Count Values')
```

```python
print("The Number of Samples in the dataset: ", len(df))
print('Class 0:', round(df['Class'].value_counts()[0]/ len(df) * 100, 2), '% ')
 
print('Class 1:', round(df['Class'].value_counts()[1]/ len(df) * 100, 2), '% ')
```

```python
Xdata = df.iloc[:, :-1]
Ydata = df.iloc[:, -1:]
 
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter
import pandas as pd
```

```python
sm = SMOTE(random_state=45)
 
X, Y = sm.fit_resample(Xdata, Ydata)
 
Xres = pd.DataFrame(X)
Yres = pd.DataFrame(Y)

 
print("After SMOTE Over Sampling Of Minor")
print("Class Total Samples are :", len(Yres))
print('Class 0 :', round(Yres.value_counts() [0] / len(Yres) * 100, 2), '% of the dataset')
print('Class 1 :', round(Yres.value_counts()[1] / len(Yres) * 100, 2), '% of the dataset')
```

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss()

Xdata = df.iloc[:, :-1]
Ydata = df.iloc[:, -1:]
x_nm, y_nm = nm.fit_resample(Xdata, Ydata)

print('Nearmiss')
print('Original dataset shape:', Counter(Y))
print('Resample dataset shape:', Counter(y_nm))
```

```python
from sklearn import preprocessing
X=Xres[['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount']].values
X=preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
y=Yres['Class']
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=45)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
from sklearn.model_selection import train_test_split

rfc = RandomForestClassifier()

rfc.fit(X_train, y_train)

# predict
rfc_predict = rfc.predict(X_test)# check performance
print('ROCAUC score:',roc_auc_score(y_test, rfc_predict))
print('Accuracy score:',accuracy_score(y_test, rfc_predict))
print('F1 score:',f1_score(y_test, rfc_predict))
```
